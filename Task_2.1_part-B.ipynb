```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
titanic = sns.load_dataset("titanic")

titanic.head()
```

```python
df = titanic[['age', 'fare']].dropna()

X = df[['age']]
y = df['fare']

print(df.shape)
```

```python
plt.figure(figsize=(8,5))

plt.scatter(X, y, alpha=0.6)

plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Titanic Dataset: Age vs Fare")

plt.show()
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)
```

```python
model_deg1 = Pipeline([
    ('poly', PolynomialFeatures(degree=1)),
    ('linear', LinearRegression())
])

model_deg1.fit(X_train, y_train)
```

```python
model_deg4 = Pipeline([
    ('poly', PolynomialFeatures(degree=4)),
    ('linear', LinearRegression())
])

model_deg4.fit(X_train, y_train)
```

```python
model_deg15 = Pipeline([
    ('poly', PolynomialFeatures(degree=15)),
    ('linear', LinearRegression())
])

model_deg15.fit(X_train, y_train)
```

```python
x_plot = np.linspace(
    X.min(),
    X.max(),
    500
).reshape(-1,1)
```

```python
y_deg1 = model_deg1.predict(x_plot)
y_deg4 = model_deg4.predict(x_plot)
y_deg15 = model_deg15.predict(x_plot)
```

```python
plt.figure(figsize=(12,7))

plt.scatter(
    X_train,
    y_train,
    color='black',
    alpha=0.5,
    label='Training Data'
)

plt.plot(
    x_plot,
    y_deg1,
    label='Degree 1',
    linewidth=3
)

plt.plot(
    x_plot,
    y_deg4,
    label='Degree 4',
    linewidth=3
)

plt.plot(
    x_plot,
    y_deg15,
    label='Degree 15',
    linewidth=3
)

plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Polynomial Regression: Titanic Dataset")

plt.legend()
plt.show()
```

```python
degrees = range(1, 21)

train_errors = []
validation_errors = []
```

```python
for degree in degrees:

    model = Pipeline([
        ('poly', PolynomialFeatures(degree=degree)),
        ('linear', LinearRegression())
    ])

    model.fit(X_train, y_train)

    train_pred = model.predict(X_train)
    val_pred = model.predict(X_val)

    train_mse = mean_squared_error(
        y_train,
        train_pred
    )

    val_mse = mean_squared_error(
        y_val,
        val_pred
    )

    train_errors.append(train_mse)
    validation_errors.append(val_mse)
```

```python
plt.figure(figsize=(10,6))

plt.plot(
    degrees,
    train_errors,
    marker='o',
    label='Training Error'
)

plt.plot(
    degrees,
    validation_errors,
    marker='o',
    label='Validation Error'
)

plt.xlabel("Polynomial Degree")
plt.ylabel("Mean Squared Error")

plt.title("Overfitting Demonstration")

plt.legend()
plt.grid(True)

plt.show()
```
