```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Sample data (x, y coordinates)
X = [
    [1, 2],
    [2, 5],
    [5, 3],
    [8, 4],
    [6, 9],
    [10, 8]
]

# Create K-Means model with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=42)

# Train the model
kmeans.fit(X)

# Get cluster labels
labels = kmeans.labels_

# Print results
print("Cluster Labels:", labels)

# Plot the points
for i in range(len(X)):
    plt.scatter(X[i][0], X[i][1], c=f"C{labels[i]}")

# Plot cluster centers
centers = kmeans.cluster_centers_
plt.scatter(
    centers[:, 0],
    centers[:, 1],
    color="black",
    marker="X",
    s=200,
    label="Centers"
)

plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans

# Student marks
marks = [[10], [12], [15], [50], [55], [60],[80],[85],[90]]

# Create K-Means model with 2 groups
model = KMeans(n_clusters=3, random_state=1)

# Train the model
model.fit(marks)

# Display each student's group
for i in range(len(marks)):
    print("Marks:", marks[i][0], "Group:", model.labels_[i])

    
```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Sample data (x, y coordinates)
X = [
    [1, 1],
    [1, 2],
    [2, 1],
    [8, 8],
    [8, 9],
    [9, 8],
    [15, 15]   # This point is far away (noise)
]

# Create DBSCAN model
model = DBSCAN(eps=2, min_samples=3)

# Train the model
model.fit(X)

# Print cluster labels
print(model.labels_)

# Draw each point
for i in range(len(X)):
    x = X[i][0]
    y = X[i][1]

    if model.labels_[i] == -1:
        plt.scatter(x, y, color="black", marker="x", s=100)
    else:
        plt.scatter(x, y, c=f"C{model.labels_[i]}", s=100)

plt.title("DBSCAN Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Sample data
X = [
    [1, 2],
    [2, 2],
    [2, 4],
    [8, 8],
    [8, 9],
    [25, 25]
]

# Create DBSCAN model
model = DBSCAN(eps=2, min_samples=2)

# Train the model
model.fit(X)

# Print cluster labels
print(model.labels_)

# Draw each point
for i in range(len(X)):
    x = X[i][0]
    y = X[i][1]
    label = model.labels_[i]

    plt.scatter(x, y, label=label)

# Show graph
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Input
X = [[1], [2], [3], [4], [5]]

# Output
y = [30, 40, 50, 60, 70]

# Train model
model = LinearRegression()
model.fit(X, y)

# Predict values for the line
predicted = model.predict(X)


# Draw original points
plt.scatter(X, y, color="blue")

# Draw regression line
plt.plot(X, predicted, color="red")

plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Simple Linear Regression")
plt.show()
```
