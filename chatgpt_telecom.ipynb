```python
# ==========================================
# STEP 1 - Import Libraries
# ==========================================

import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

import warnings
warnings.filterwarnings("ignore")


# ==========================================
# STEP 2 - Load Dataset
# ==========================================

df = pd.read_csv("telecom_churn_mock_data.csv")

print(df.shape)
print(df.head())


# ==========================================
# STEP 3 - Remove CustomerID
# ==========================================

df.drop("CustomerID", axis=1, inplace=True)


# ==========================================
# STEP 4 - Encode Binary Columns
# ==========================================

df['Gender'] = df['Gender'].map({'Male':0,'Female':1})

df['Partner'] = df['Partner'].map({'No':0,'Yes':1})

df['Dependents'] = df['Dependents'].map({'No':0,'Yes':1})

df['PhoneService'] = df['PhoneService'].map({'No':0,'Yes':1})

df['PaperlessBilling'] = df['PaperlessBilling'].map({'No':0,'Yes':1})

df['Churn'] = df['Churn'].map({'No':0,'Yes':1})


# ==========================================
# STEP 5 - Encode Remaining Categorical Columns
# ==========================================

df['MultipleLines'] = df['MultipleLines'].map({
    'No':0,
    'Yes':1,
    'No phone service':2
})

df['InternetService'] = df['InternetService'].map({
    'DSL':0,
    'Fiber optic':1,
    'No':2
})

df['OnlineSecurity'] = df['OnlineSecurity'].map({
    'No':0,
    'Yes':1,
    'No internet service':2
})

df['OnlineBackup'] = df['OnlineBackup'].map({
    'No':0,
    'Yes':1,
    'No internet service':2
})

df['DeviceProtection'] = df['DeviceProtection'].map({
    'No':0,
    'Yes':1,
    'No internet service':2
})

df['TechSupport'] = df['TechSupport'].map({
    'No':0,
    'Yes':1,
    'No internet service':2
})

df['StreamingTV'] = df['StreamingTV'].map({
    'No':0,
    'Yes':1,
    'No internet service':2
})

df['StreamingMovies'] = df['StreamingMovies'].map({
    'No':0,
    'Yes':1,
    'No internet service':2
})

df['Contract'] = df['Contract'].map({
    'Month-to-month':0,
    'One year':1,
    'Two year':2
})

df['PaymentMethod'] = df['PaymentMethod'].map({
    'Electronic check':0,
    'Mailed check':1,
    'Bank transfer (automatic)':2,
    'Credit card (automatic)':3
})


# ==========================================
# STEP 6 - Check Missing Values
# ==========================================

print(df.isnull().sum())

# Remove rows with missing values if any
df.dropna(inplace=True)


# ==========================================
# STEP 7 - Select Features
# ==========================================

X = df.drop("Churn", axis=1)

y = df["Churn"]


# ==========================================
# STEP 8 - Split Dataset
# ==========================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)


# ==========================================
# STEP 9 - Train Random Forest
# ==========================================

model = RandomForestClassifier(
    n_estimators=200,
    random_state=42
)

model.fit(X_train, y_train)

print("Model Trained Successfully")


# ==========================================
# STEP 10 - Predict
# ==========================================

y_pred = model.predict(X_test)


# ==========================================
# STEP 11 - Accuracy
# ==========================================

accuracy = accuracy_score(y_test,y_pred)

print("Accuracy :",accuracy*100)


# ==========================================
# STEP 12 - Classification Report
# ==========================================

print(classification_report(
    y_test,
    y_pred,
    target_names=["Stayed","Churned"]
))


# ==========================================
# STEP 13 - Feature Importance
# ==========================================

importance = pd.DataFrame({

    "Feature":X.columns,

    "Importance":model.feature_importances_

})

importance = importance.sort_values(
    by="Importance",
    ascending=False
)

print("\nTop Important Features\n")

print(importance)


# ==========================================
# STEP 14 - Train Again Using Top Features
# ==========================================

top_features = importance.head(8)["Feature"].tolist()

print("\nSelected Features")

print(top_features)

X = df[top_features]

y = df["Churn"]


X_train,X_test,y_train,y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

model = RandomForestClassifier(
    n_estimators=200,
    random_state=42
)

model.fit(X_train,y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test,y_pred)

print("\nAccuracy Using Top Features")

print(f"{accuracy*100:.2f}%")

print(classification_report(
    y_test,
    y_pred,
    target_names=["Stayed","Churned"]
))
```
